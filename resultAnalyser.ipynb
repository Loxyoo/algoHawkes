```python
import struct
for sym in ['ETHUSD','BNBUSD','SOLUSD','XRPUSD']:
    with open(f'optimized_params_{sym}.bin','rb') as f:
        data = f.read()
    off = 0
    status = struct.unpack_from('i', data, off)[0]; off += 4
    sym_len = struct.unpack_from('Q', data, off)[0]; off += 8
    symbol = data[off:off+sym_len].decode(); off += sym_len
    def read_vec(off):
        n = struct.unpack_from('Q', data, off)[0]; off += 8
        vec = struct.unpack_from(f'{n}d', data, off); off += 8*n
        return list(vec), off
    alpha, off = read_vec(off)
    beta,  off = read_vec(off)
    mu,    off = read_vec(off)
    phi,   off = read_vec(off)
    n = int(len(alpha)**0.5)
    print(f'=== {symbol} ===')
    print('phi matrix:')
    for i in range(n):
        print('  ', [f'{phi[i*n+j]:.3g}' for j in range(n)])
    print("Alpha matrix")
    for i in range(n):
        print(' ', [f'{alpha[i*n+j]:.3g}' for j in range(n)])
    print("Beta matrix")
    for i in range(n):
        print(' ', [f'{beta[i*n+j]:.3g}' for j in range(n)])
    print("mu vector")
    print([mu[i] for i in range(n)])
    print()
```
